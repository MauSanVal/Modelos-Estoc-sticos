## Importar librerías

```python
import sympy as sp
import networkx as nx
import matplotlib.pyplot as plt
```

## Creación de matrices

```python
A = sp.Matrix([[1,2],[3,4]])
B = sp.Matrix([[5,6],[7,8]])
sp.init_printing()
display(A,B)
```

## Acceder a un elemento de una matriz

```python
#Acceder a la entrada 1,1 de la matriz A (los índices comienzan en 0)
A11 = A[1,1]
print(A11)
```

## Multiplicación matriz-vector

```python
v = sp.Matrix([[1],[2]])
A*v
```

## Multiplicación de matrices

```python
A*B
```

## Potencias de matrices

```python
A**2
A**3
```

## Resolver sistemas lineales

```python
A = sp.Matrix([
    [1,1,1],
    [2,-1,1],
    [1,2,3]
])

b = sp.Matrix([
    [6],
    [3],
    [14]
])

sol = A.LUsolve(b)

sol
```

## Valores propios y vectores propios

```python
M = sp.Matrix([
    [4,1,1],
    [1,4,1],
    [1,1,4]
])

print("Matriz:")
display(M)

print("Valores propios:")
display(M.eigenvals())

print("Vectores propios:")
display(M.eigenvects())
```

## Crear diagrama de transición con NetworkX

```python
G = nx.DiGraph()
G.add_nodes_from(['A','B','C'])
G.add_edge('A','B')
G.add_edge('A','C')
G.add_edge('B','C')
G.add_edge('C','A')
```

## Dibujar el grafo

```python
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightblue')
plt.show()
```

## Diagrama de transición con pesos

```python
G = nx.DiGraph()
G.add_weighted_edges_from([
('A','B',0.5),
('A','C',0.5),
('B','C',1.0),
('C','A',1.0)
])

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightgreen', connectionstyle = 'arc3, rad = 0.3')

labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, label_pos = 0.3)

plt.show()
```
