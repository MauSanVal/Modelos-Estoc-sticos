# Caminatas al azar

Tenemos un punto que se puede movera la derecha con probabilidad $p$ y a la izquierda con probabilidad $q$, tal que $p+q=1$

```python
import random as rd
```

```python
p = 0.5
#Por cómo se escribe el código, necesitamos solamente el valor de p, se entiende que q es el complemento a 1 de p
```

```python
def mov():
    if rd.random() <= p:
        return +1
    else:
        return -1
```

```python
n = 1000000
```

```python
posicion = [0]*n
posicion[0] = 0
```

```python
for i in range(1,n):
    posicion[i] = posicion[i-1] + mov()
```

```python
import matplotlib.pyplot as pt
```

```python
pt.plot(posicion)
pt.title("Caminata aleatoria para p=0.5")
pt.xlabel("Número de movimientos")
pt.ylabel("Posición")
pt.axhline(y=0)
```

```python
#Encontrando la posición máxima y mínima
maximo = max(posicion)
minimo = min(posicion)
print("El máximo es: ",maximo)
print("El mínimo es: ",minimo)
```

```python
#Buscando la cantidad de veces que se pasa por el 0
print("Se pasa por el cero: ",posicion.count(0) - 1," veces") #se resta uno porque se incia en 0
```

```python
#Viendo cuánto tarda en regresar a 0 por primera vez
print("Se regresó al cero por primera vez después de: ", posicion.index(0,1), " movimientos")
```

```python
suma = 0
for i in range(100):
    suma_aux = 0
    for i in range(1000000):
        suma_aux += mov()
    suma += suma_aux
promedio = suma/100
print("La posición promedio después de n=1000000 de movimientos es: ", promedio)
```
