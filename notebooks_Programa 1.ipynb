```python
import random as rd
```

```python
def volado():
    return round(rd.random())
```

```python
volado()
```

```python
n = 1000000
```

```python
suma = 0
sumas = [0]*n
```

```python
for i in range(1,n):
    suma += volado()
    sumas[i] = suma/i
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(sumas)
```
