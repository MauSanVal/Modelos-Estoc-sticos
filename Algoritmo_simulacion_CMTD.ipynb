Tenemos un tablero de $4$ sectores, lanzamos $2$ monedas y avanzamos en sentido horario, el número de caras que obtengamos. Si repetimos el procedimiento ¿son todos los sectores igual de probables cuando pase mucho timepo?
(Los colores están en el siguiente orden en sentido horario: rojo, naranja, azul, verde).
Tenemos el conjunto de estados $S={0,1,2,3}$ , donde $0=rojo , 1=naranja , 2=azul , 3=verde$

```python
import numpy as np
import random
```

```python

#Estado inicial
alfa = [1/4, 1/4, 1/4, 1/4]
#Matriz de transición de un paso
P = np.array([
    [1/4, 1/2, 1/4, 0],
    [0, 1/4, 1/2, 1/4],
    [1/4, 0, 1/4, 1/2],
    [1/2, 1/4, 0, 1/4]
])
```

```python
#Función que genera el estado X_i+1 a partir del estado X_i
def Xi(p):
    r = random.random()
    acum = 0
    for i, prob in enumerate(p):
        acum += prob
        if r <= acum:
            return i
```

```python
#Escogiendo estado X_0
resultados = []
resultados.append(Xi(alfa))
```

```python
#Ciclo para X_i i>=2
for i in range(100000):
    resultados.append(Xi(P[resultados[i]]))
```

```python
print(resultados)
```

```python
#Analizando si se cumple la distribución teórica equiprobable a 1/4
print("Probabilidad de que caiga naranja: ", resultados.count(0)/len(resultados))
print("Probabilidad de que caiga azul: ", resultados.count(1)/len(resultados))
print("Probabilidad de que caiga verde: ", resultados.count(2)/len(resultados))
print("Probabilidad de que caiga rojo: ", resultados.count(3)/len(resultados))
```

```python
#Vemos que las probabilidades son muy cercanas a 1/4, que es la probabilidad teórica, por lo tanto es una buena simulación del sistema.
```
